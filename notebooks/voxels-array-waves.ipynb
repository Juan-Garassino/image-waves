```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product

STYLE = "#ffffff"
```

```python
# volumes = np.random.randint(2, size=(28,28,28))

volumes = np.random.choice(2, size=(28,28,28), p=(0.8, 0.2))
```

```python
fig, axes = plt.subplots(ncols=5, nrows=6, figsize=(25, 25), facecolor = (STYLE))
axes = axes.ravel()

for index in range(28):
    axes[index].imshow(volumes[index], cmap = "plasma")
```
